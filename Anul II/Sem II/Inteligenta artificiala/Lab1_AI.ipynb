```python
# Problema 1

# Complexitate - n este lungimea sirului dat:
# timp: O(n)
# spatiu: O(n)

def ultimul_cuvant(sir):

  """
  Determina ultimul cuvant din punct de vedere lexicografic dintr-un sir dat.
  :param sir: Sirul initial.
  :return: Ultimul cuvant din punct de vedere lexicografic.
  """

  cuvinte = sir.split()
  cuvant_salvat = 'a'
  for cuvant in cuvinte:
    if cuvant[0] >= cuvant_salvat[0]:
      cuvant_salvat = cuvant
  return cuvant_salvat


test1 = ultimul_cuvant("Ana are 5 mere rosii si galbene")
test2 = ultimul_cuvant("Ana are mere rosii si galbene si un zmeu")

assert test1 == "si"
assert test2 == "zmeu"


#AI generate code - chatgpt

#Complexitate:
# Time Complexity: O(n), where n is the number of words in the text
# (splitting and max operations are linear).
# Space Complexity: O(n), as we store the words in a list.

def ultimul_cuvant_alfabetic(text):
    cuvinte = text.split()

    ultimul_cuvant = max(cuvinte)

    return ultimul_cuvant


# Ambele complexitati, atat cea generata de bot inteligent cat si cea
# implementata de mine este identica, atat complexitatea timp cat si cea
# de spatiu esta una liniara, O(n) unde n este lungimea sirului initial

```

```python
#Problema 2

#Complexitate:
# Timp: O(1) - se realizeaza in timp constant fiind executat pe valori date si fixe
# Spatiun: O(1)

import math


def distanta_euclidiana(x1, y1, x2, y2):
  """
  Calculate the Euclidean distance between two points.
  :param x1: x-coordinate of the first point
  :param y1: y-coordinate of the first point
  :param x2: x-coordinate of the second point
  :param y2: y-coordinate of the second point
  :return: Euclidean distance
  """
  return math.sqrt(((x1-x2)**2+(y1-y2)**2))

test1 = distanta_euclidiana(1,5,4,1)
test2 = distanta_euclidiana(2,10,8,2)

assert test1 == 5.0
assert test2 == 10.0

#AI generate code - chatgpt

# Complexitate:
# Time Complexity: O(1), as we perform a constant number of operations.
# Space Complexity: O(1), as we use a fixed amount of space.

import math

def distanta_euclidiana(p1, p2):
    return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)



```

```python
#Problema 3

#compelexitate
# de timp: O(n) se foloseste un singur contor
# de spatiu: O(1) stocam o valoare numerica

def produs_scalar2(v1,v2):
  produs = 0
  i = 0
  while i < len(v1):
    if v1[i] != 0 and v2[i] != 0:
      produs += v1[i]*v2[i]
    i += 1
  return produs

assert produs_scalar2([1,0,2,0,3],[1,2,0,3,1]) == 4
assert produs_scalar2([2,0,1,9,0],[1,2,3,0,1]) == 5


#AI generate

#compelexitate
# de timp: O(n) se foloseste functia zip pentru a parcurge in paralel cei doi vectori
# de spatiu: O(1) stocam o valoare numerica

def sparse_dot_product(vec1, vec2):
    if len(vec1) != len(vec2):
        raise ValueError("Vectorii trebuie să aibă aceeași dimensiune.")

    dot_product = sum(v1 * v2 for v1, v2 in zip(vec1, vec2) if v1 != 0 and v2 != 0)
    return dot_product

# Exemplu de utilizare
vec1 = [1, 0, 2, 0, 3]
vec2 = [1, 2, 0, 3, 1]

assert sparse_dot_product(vec1,vec2) == 4
```

```python
#Problema 4

#Complexitat:
# timp: O(n)
# spatiu: O(n)

def singura_aparitie(sir):
  """
  Suprogramul determina cuvintele ce apar o singura data intr-un sir
  :param sir: string, sir de caractere dat
  :return: string, sirul format cu cuvintele ce apar o singura data
  """
  rez = ""
  cuvinte = sir.split()
  for cuvant in cuvinte:
    if cuvinte.count(cuvant) == 1:
      rez += cuvant
      rez += " "
  return rez.strip()

test1 = singura_aparitie("ana are ana are mere rosii ana")
assert test1 == "mere rosii"

#AI generate

from collections import Counter

def unique_words(text):
    words = text.split()
    word_counts = Counter(words)

    unique = [word for word, count in word_counts.items() if count == 1]

    return unique

# Time Complexity: O(n), where n is the number of words in the text (due to splitting, counting, and filtering operations).
# Space Complexity: O(n), as we store the word counts in a dictionary and the unique words in a list.

```

```python
#Problema 5

#Complexitate:
# timp: O(n^2)
# spatiu: O(1)

def valoare_duplicat(v):
  """
  Suprogramul valorile duplicate dintr-un vector
  :param integer v: sirul de numere de la 1 la n-1, unde n este lungimea
  vectorului
  :return: integer valoare care apare de doua ori
  """
  n = len(v)
  for i in range(n):
    for j in range(i+1,n):
      if v[i] == v[j]:
        return v[i]

test1 = valoare_duplicat([1,2,3,4,2])
assert test1 == 2


#AI generated

# Time Complexity: O(n), as we iterate through the list once.
# Space Complexity: O(n), since we store seen elements in a set.

def find_duplicate(arr):
    seen = set()
    for num in arr:
        if num in seen:
            return num
        seen.add(num)
    return None





```

```python
#Problema 6

#Complexitate:
# timp: O(n)
# spatiu: O(n)

def element_majoritar(v):
  """
  Suprogramul determina elementul majoritar dintr-un vector
  :param integer v: vector de numere intregi
  :return: integer elementul care apare de cel putin n/2 ori unde n este
  lungimea vectorului
  """
  n = len(v)
  elements = [0]*n
  for elem in v:
    elements[elem] += 1
  for i in range(len(elements)):
    if elements[i] >= n/2:
      return i


test1 = element_majoritar([2,8,7,2,2,5,2,3,1,2,2])
assert test1 == 2

#AI generate

from collections import Counter

def find_majority_element(arr):
    n = len(arr)
    counts = Counter(arr)
    for num, count in counts.items():
        if count > n // 2:
            return num
    return None  # No majority element found

# Example usage
arr = [2, 8, 7, 2, 2, 5, 2, 3, 1, 2, 2]
print(find_majority_element(arr))  # Output: 2

# Time Complexity: O(n), as we count elements in one pass and check in another.
# Space Complexity: O(n), since we store counts in a dictionary.
```

```python
#Problema 7

#Complexitate
# timp: O(nlogn)
# spatiu: O(1)

def al_k_cel_mai_mare(k,v):
  """
  Suprogramul determina al k lea cel mai mare element dintr-un vector
  :param integer v: vector de numere intregi
  :param integer k: pozitia elementului
  :return: integer al k lea element maxim
  """
  v.sort(reverse=True)
  return v[k-1]

test1 = al_k_cel_mai_mare(2,[7,4,6,3,9,1])
test2 = al_k_cel_mai_mare(3,[7,4,6,3,9,1])
assert test1 == 7
assert test2 == 6

#AI generate

import heapq

def al_k_cel_mai_mare_optim(k, v):
    # Construim un heap de dimensiune k
    heap = v[:k]  # Luăm primele k elemente
    heapq.heapify(heap)  # Le transformăm într-un heap min

    # Parcurgem restul elementelor
    for num in v[k:]:
        if num > heap[0]:  # Dacă elementul curent este mai mare decât cel mai mic din heap
            heapq.heappushpop(heap, num)  # Adăugăm și eliminăm minimul

    return heap[0]  # Cel mai mic element din heap este al k-lea cel mai mare

# Complexitate:
# - Timp: O(n log k) deoarece construirea heapului durează O(k), iar fiecare inserare/extragere durează O(log k) și se face pentru n-k elemente.
# - Spațiu: O(k) deoarece păstrăm doar k elemente în heap.
```

```python
#Problema 8

#Complexitate:
# timp: O(nlogn)
# spatiu: O(nlogn)

def nr_binare(n):
  """
  Suprogramul determina toate numerele binare de la 1 la n
  :param integer n: numarul intreg n
  :return v: sirul de numere binare de la 1 la n
  """
  v = []
  i = 1
  while i <= n:
    binary = bin(i)[2:]
    v.append(binary)
    i += 1
  return v

test1 = nr_binare(4)
assert test1 == ['1', '10', '11', '100']


#AI generate

#Complexitate:
# timp: O(nlogn)
# spatiu: O(nlogn)

def nr_binare(n):
    v = []
    for i in range(1, n + 1):  # Iterate from 1 to n
        binary = bin(i)[2:]  # Convert i to binary, removing '0b' prefix
        v.append(binary)  # Store the binary string

    return " ".join(v)
```

```python
#Problema 9

# Complexitate
# timp: O(n*m)
# spatiu: O(1)

def suma_subm(matrice,p,q,r,s):
  """
  Suprogramul determina suma elementelor dintr-o submatrice
  :param matrice: matrice de numere intregi
  :param p: linia de inceput a submatricei
  :param q: coloana de inceput a submatricei
  :param r: linia de sfarsit a submatricei
  :param s: coloana de sfarsit a
  :return: suma elementelor din submatrice
  """
  n = len(matrice)
  m = len(matrice[0])
  sum = 0
  for i in range (0,n):
    for j in range (0,m):
      if i >= p and i <= r and j >= q and j <= s:
        sum += matrice[i][j]
  return sum

def suma_p(matrice, lista):
  """
  Suprogramul determina suma elementelor dintr-o submatrice
  :param matrice: matrice de numere intregi
  :param lista: lista de submatrici
  :return: suma elementelor din submatrice
  """
  rez = []
  for submatrice in lista:
    p,q,r,s = submatrice
    rez.append(suma_subm(matrice,p,q,r,s))
  return rez

test1 = suma_subm([[0, 2, 5, 4, 1],
[4, 8, 2, 3, 7],
[6, 3, 4, 6, 2],
[7, 3, 1, 8, 3],
[1, 5, 7, 9, 4]],1,1,3,3)

test2 = suma_subm([[0, 2, 5, 4, 1],
[4, 8, 2, 3, 7],
[6, 3, 4, 6, 2],
[7, 3, 1, 8, 3],
[1, 5, 7, 9, 4]],2,2,4,4)

lista = [(1, 1, 3, 3), (2, 2, 4, 4)]

test11 = suma_p([[0, 2, 5, 4, 1],
[4, 8, 2, 3, 7],
[6, 3, 4, 6, 2],
[7, 3, 1, 8, 3],
[1, 5, 7, 9, 4]],lista)

assert test11 == [38,44]

#AI generate

def sum_submatrices(matrix, pairs):
    """
    Calculate the sum of elements in submatrices defined by given pairs of coordinates.
    """
    prefix_sum = [[0] * (len(matrix[0]) + 1) for _ in range(len(matrix) + 1)]

    # Compute the prefix sum matrix
    for i in range(1, len(matrix) + 1):
        for j in range(1, len(matrix[0]) + 1):
            prefix_sum[i][j] = (
                matrix[i-1][j-1] +
                prefix_sum[i-1][j] +
                prefix_sum[i][j-1] -
                prefix_sum[i-1][j-1]
            )

    results = []
    for (p, q), (r, s) in pairs:
        total = (
            prefix_sum[r+1][s+1]
            - prefix_sum[p][s+1]
            - prefix_sum[r+1][q]
            + prefix_sum[p][q]
        )
        results.append(total)

    return results

# Example usage
matrix = [
    [0, 2, 5, 4, 1],
    [4, 8, 2, 3, 7],
    [6, 3, 4, 6, 2],
    [7, 3, 1, 8, 3],
    [1, 5, 7, 9, 4]
]

pairs = [((1, 1), (3, 3)), ((2, 2), (4, 4))]

result = sum_submatrices(matrix, pairs)
print(result)  # Output: [38, 44]

"""
Time Complexity: O(n * m) for preprocessing, O(1) per query.
Space Complexity: O(n * m) for storing prefix sums.
"""
```

```python
#Problema 10

#Complexitate:
#timp: O(n*m)
#spatiu: O(1)

def maxim_1(n,m, matrice):
  """
  Suprogramul determina linia cu cele mai multe valori de 1
  :param integer n: numarul de linii
  :return v: sirul de numere binare de la 1 la n
  """
  maxim = 0
  ind_i = -1
  for i in range(0,n):
    sum = 0
    for j in range(0,m):
      sum += matrice[i][j]
    if(sum > maxim):
      maxim = sum
      ind_i = i
  return ind_i+1


test1 = maxim_1(3,5,[[0,0,0,1,1],
[0,1,1,1,1],
[0,0,1,1,1]])
assert test1 == 2

#Ai generate

def linie_max_1(matrice):
    n = len(matrice)    # Numărul de linii
    m = len(matrice[0]) # Numărul de coloane
    max_ones = 0        # Numărul maxim de 1
    index_linie = -1    # Indexul liniei cu cei mai mulți 1

    for i in range(n):
        # Numărăm câți de 1 sunt în linia curentă
        ones_count = m - binary_search_first_one(matrice[i])

        if ones_count > max_ones:
            max_ones = ones_count
            index_linie = i

    return index_linie

def binary_search_first_one(row):
    """Găsește primul 1 într-un rând sortat folosind căutare binară."""
    left, right = 0, len(row) - 1
    while left <= right:
        mid = (left + right) // 2
        if row[mid] == 1:
            right = mid - 1  # Căutăm mai la stânga
        else:
            left = mid + 1   # Mergem mai la dreapta
    return left  # Indicele primului 1 sau m dacă nu există

# Complexitate:
# - Timp: O(n log m), deoarece parcurgem fiecare linie (O(n)) și aplicăm căutare binară (O(log m)).
# - Spațiu: O(1), deoarece folosim doar câțiva scalari suplimentari.
```
